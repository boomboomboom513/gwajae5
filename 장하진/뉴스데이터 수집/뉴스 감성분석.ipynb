```python
# 감성분석 모델 불러오기
# AutoTokenizer -> Tokenizer로 입력된 데이터를 token으로 바꾸는 역할
# AutoModelForSequenceClassification은 Hugging Face 라이브러리의 Transformers에서 제공하는 클래스이며
# 자동으로 시퀀스 분류 작업에 사용할 수 있는 사전 훈련된 모델을 선택하고, 이를 미세 조정(fine-tuning)하는 데 사용한다.
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline
import pandas as pd
```

```python
tokenizer = AutoTokenizer.from_pretrained('snunlp/KR-FinBert-SC')
model = AutoModelForSequenceClassification.from_pretrained('snunlp/KR-FinBert-SC')
classifier = pipeline(task='text-classification', model=model, tokenizer=tokenizer)
```

```python
newsData = pd.read_excel('./NewsResult_20220101-20230704.xlsx')
newsData.tail(10)
```

```python
# neutral 중립, positive 긍정, negative 부정
if classifier(newsData['본문'][30])[0]['label'] == 'positive':
    print('이 뉴스는 긍정적 뉴스이다. 긍정적일 확률 : '+str(round(classifier(newsData['본문'][30])[0]['score'] * 100, 2))+'%')
elif classifier(newsData['본문'][30])[0]['label'] == 'negative':
    print('이 뉴스는 부정적 뉴스이다. 부정적일 확률 : '+str(round(classifier(newsData['본문'][30])[0]['score'] * 100, 2))+'%')
else:
    print('이 뉴스는 중립적 뉴스이다. 중립적일 확률 : '+str(round(classifier(newsData['본문'][30])[0]['score'] * 100, 2))+'%')
```
