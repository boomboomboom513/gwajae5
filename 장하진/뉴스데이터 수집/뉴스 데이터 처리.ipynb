```python
import pandas as pd
import numpy as np
from collections import Counter
import re
```

```python
aaa = pd.read_csv('../뉴스데이터 수집/AI 뉴스 데이터10.csv', encoding='utf-8')
abc = aaa.copy()
abc
```

```python
word_list = []

for i in range(0, abc.shape[0]):
    content_str = str(abc.iloc[i]['내용'])
    for word in content_str.split():
        word_list.append(word)
```

```python
word_list_count = Counter(word_list)
word_list_count
```

```python
zzz = pd.DataFrame([dict(word_list_count)])
zzz
```

```python
word_df = pd.DataFrame({
      '단어':zzz.columns
    , '개수':zzz.values[0]
})
word_df
```

```python
word_df.to_csv('./AI 주식 키워드10.csv', encoding='utf-8', index=False)
```
